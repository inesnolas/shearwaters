```python
import numpy as np
import os
import librosa
from tqdm.notebook import tqdm
import csv
import json
```

```python
Report = {}
```

```python
data_processed = '/home/ines/Dropbox/QMUL/PHD/manx_shearwaters/data/data_processed/'

raw_annotations_path ='/home/ines/Dropbox/QMUL/PHD/manx_shearwaters/data/Annotations/'

labels_dict = {1:'male_in', 2:'female_in', 3:'chick', 4:'male_bout', 5:'female_bout', 6:'flapping', 7: 'male_grunt', 8:'female_grunt', 9:'unk_grunt' }
```

```python
list_wav_files = os.listdir(data_processed)
n_files = len(list_wav_files)-1
print('n_files', str(n_files))
Report['n_files'] = n_files
```

```python
# n_burrows:

burrows_ids = []
for filename in list_wav_files:
    burrow_id = filename.split('_')[0]
    burrows_ids.append(burrow_id)
    
burrows_ids = set(burrows_ids)
n_burrows = len(burrows_ids)
print('n_burrows', n_burrows)
```

```python
dataset = np.load(data_processed+'dataset_25_03.npy', allow_pickle=True)
```

```python

print('dataset shape: ', dataset.shape)
n_segments = dataset.shape[0]
print('n_segments', n_segments)
Report['n_segments']= n_segments
```

```python
# n_segments per burrow:
# n_negative per burrow
# n_segm_per_class_per_burrow

# 1:'male_in', 2:'female_in', 3:'chick', 4:'male_bout', 5:'female_bout', 6:'flapping', 7: 'male_grunt', 8:'female_grunt', 9:'unk_grunt' 
burrows_dict = {}
for segm in dataset:
    burrow_id = segm[0].split('_')[0]
    
    label_matrix = segm[-1]
    
    if burrow_id in burrows_dict.keys():
        burrows_dict[burrow_id]['n_segments'] = burrows_dict[burrow_id]['n_segments']+1
        if sum(sum(label_matrix)) == 0:
            burrows_dict[burrow_id]['n_neg_segments'] = burrows_dict[burrow_id]['n_neg_segments']+1
        active_classes = np.where(np.sum(label_matrix,1) > 0 )
#         print(active_classes)
#         print(np.sum(label_matrix,1))
         # by hand :(
        for i in active_classes[0]:
#             print(i)
            if i ==0:
                burrows_dict[burrow_id]['n_seg_male_in'] = burrows_dict[burrow_id]['n_seg_male_in'] + 1
            if i ==1:
                 burrows_dict[burrow_id]['n_seg_female_in'] = burrows_dict[burrow_id]['n_seg_female_in'] + 1
            if i ==2:
                 burrows_dict[burrow_id]['n_seg_chick'] = burrows_dict[burrow_id]['n_seg_chick'] + 1
         
            if i ==3:
                burrows_dict[burrow_id]['n_seg_male_bout'] = burrows_dict[burrow_id]['n_seg_male_bout'] + 1
            if i ==4:
                 burrows_dict[burrow_id]['n_seg_female_bout'] = burrows_dict[burrow_id]['n_seg_female_bout'] + 1
         
            if i ==5:
                 burrows_dict[burrow_id]['n_seg_flapping'] = burrows_dict[burrow_id]['n_seg_flapping'] + 1
         
            if i ==6:
                 burrows_dict[burrow_id]['n_seg_male_grunt'] = burrows_dict[burrow_id]['n_seg_male_grunt'] + 1
         
            if i ==7:
                 burrows_dict[burrow_id]['n_seg_female_grunt'] = burrows_dict[burrow_id]['n_seg_female_grunt'] + 1
         
            if i ==8:
                 burrows_dict[burrow_id]['n_seg_unk_grunt'] = burrows_dict[burrow_id]['n_seg_unk_grunt'] + 1
         
                
    else:
        burrows_dict[burrow_id]= {'n_segments':1, 'n_neg_segments':0, \
                                  'n_seg_male_in':0, 'n_seg_female_in':0, \
                                  'n_seg_chick':0, 'n_seg_male_bout':0,\
                                  'n_seg_female_bout':0, 'n_seg_flapping':0,\
                                  'n_seg_male_grunt':0, 'n_seg_female_grunt':0, 'n_seg_unk_grunt':0 }
        if sum(sum(label_matrix)) == 0:
            burrows_dict[burrow_id]['n_neg_segments'] = burrows_dict[burrow_id]['n_neg_segments'] +1
        active_classes = np.where(np.sum(label_matrix,1) > 0 )
#         print(active_classes)
#         print(np.sum(label_matrix,1))
         # by hand :(
        for i in active_classes[0]:
#             print(i)
            
            if i ==0:
                burrows_dict[burrow_id]['n_seg_male_in'] = burrows_dict[burrow_id]['n_seg_male_in'] + 1
            if i ==1:
                 burrows_dict[burrow_id]['n_seg_female_in'] = burrows_dict[burrow_id]['n_seg_female_in'] + 1
            if i ==2:
                 burrows_dict[burrow_id]['n_seg_chick'] = burrows_dict[burrow_id]['n_seg_chick'] +1
         
            if i ==3:
                burrows_dict[burrow_id]['n_seg_male_bout'] = burrows_dict[burrow_id]['n_seg_male_bout'] + 1
            if i ==4:
                 burrows_dict[burrow_id]['n_seg_female_bout'] = burrows_dict[burrow_id]['n_seg_female_bout'] + 1
         
            if i ==5:
                 burrows_dict[burrow_id]['n_seg_flapping'] = burrows_dict[burrow_id]['n_seg_flapping'] + 1
         
            if i ==6:
                 burrows_dict[burrow_id]['n_seg_male_grunt'] = burrows_dict[burrow_id]['n_seg_male_grunt'] + 1
         
            if i ==7:
                 burrows_dict[burrow_id]['n_seg_female_grunit'] = burrows_dict[burrow_id]['n_seg_female_grunt'] + 1
         
            if i ==8:
                 burrows_dict[burrow_id]['n_seg_unk_grunt'] = burrows_dict[burrow_id]['n_seg_unk_grunt'] + 1
         
                
            
# print(burrows_dict)
import json

print(json.dumps(burrows_dict, indent=1))
```

```python
with open("analysis_per_burrow.json", 'w') as oputfile:
    json.dump(burrows_dict, oputfile) 
```

```python
## per class:

#n_segments_per_class
#n_frames_per_class
# total_number_events, mean_duration, variance, max_duration, min_duration, \

labels_duration_dict = {'male_in':[], 'female_in':[],\
               'chick':[], 'male_bout':[], \
               'female_bout':[],'flapping':[], \
               'male_grunt':[],'female_grunt':[],\
               'unk_grunt':[]}




# go back to raw annotations to get idea of whole events characteristics
for rawfilename in os.listdir(raw_annotations_path):
    
#     print(rawfilename)
    with open(raw_annotations_path + rawfilename, 'r') as rawfile:
        reader = csv.reader(rawfile)

        for row in reader:
            try:
                label_value = int(row[1])
#                     print(row[1])
            except ValueError as e:
                print("annotation error, will make it as closest label!")
                label_value = int(np.floor(float(row[1])))

            label_str = labels_dict[int(label_value)]
            labels_duration_dict[label_str].append(float(row[2]))
#             print(labels_duration_dict[label_str])
                



```

```python
labels_sumaary_dict = {'male_in':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'female_in':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'chick':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'male_bout':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'female_bout':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'flapping':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'male_grunt':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'female_grunt':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'unk_grunt':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}}


for key in labels_duration_dict.keys():
    nb_events = len(labels_duration_dict[key])
    mean_duration = np.mean(labels_duration_dict[key])
    sd = np.std(labels_duration_dict[key])
    max_d = np.max(labels_duration_dict[key])
    min_d = np.min(labels_duration_dict[key])
    labels_sumaary_dict[key]['nb_events'] = nb_events
    labels_sumaary_dict[key]['mean_duration'] = mean_duration
    labels_sumaary_dict[key]['duration_sd'] = sd
    labels_sumaary_dict[key]['max_duration'] = max_d
    labels_sumaary_dict[key]['min_duration'] = min_d
    
print(json.dumps(labels_sumaary_dict, indent=1))

    
```

```python
with open("analysis_per_class.json", 'w') as oputfile:
    json.dump(labels_sumaary_dict, oputfile) 
```
