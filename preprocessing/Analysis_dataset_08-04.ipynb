## Numbers per class:

```python
## per class:
#n_segments_per_class
#n_frames_per_class
#total_number_events, mean_duration, variance, max_duration, min_duration, \

labels_duration_dict = {'male_in':[], 'female_in':[],\
               'chick':[], 'male_bout':[], \
               'female_bout':[],'flapping':[], \
               'male_grunt':[],'female_grunt':[],\
               'unk_grunt':[]}




# go back to raw annotations to get idea of whole events characteristics
for rawfilename in os.listdir(raw_annotations_path):
    with open(raw_annotations_path + '/' + rawfilename, 'r') as rawfile:
        reader = csv.reader(rawfile)
        for row in reader:
            try:
                label_value = int(row[1])
            except ValueError as e:
                print("annotation error, will make it as closest label!")
                label_value = int(np.floor(float(row[1])))
                print(row[1])
            label_str = labels_dict[int(label_value)]
            labels_duration_dict[label_str].append(float(row[2]))
```

```python
labels_sumaary_dict = {'male_in':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'female_in':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'chick':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'male_bout':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'female_bout':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'flapping':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}, \
                       'male_grunt':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'female_grunt':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0},\
                       'unk_grunt':{'nb_events':0, 'mean_duration':0, 'duration_sd':0, 'max_duration':0, 'min_duration':0}}


for key in labels_duration_dict.keys():
    nb_events = len(labels_duration_dict[key])
    mean_duration = np.mean(labels_duration_dict[key])
    sd = np.std(labels_duration_dict[key])
    max_d = np.max(labels_duration_dict[key])
    min_d = np.min(labels_duration_dict[key])
    labels_sumaary_dict[key]['nb_events'] = nb_events
    labels_sumaary_dict[key]['mean_duration'] = mean_duration
    labels_sumaary_dict[key]['duration_sd'] = sd
    labels_sumaary_dict[key]['max_duration'] = max_d
    labels_sumaary_dict[key]['min_duration'] = min_d
    
print(json.dumps(labels_sumaary_dict, indent=1)) 
```

## Overlaps:

```python
burrow_overlap_dictionary = {}
overlap_lengths_dict = {'total_male_length':0, 'total_female_length':0, \
                        'total_chick_length':0, 'm_f_overlap_length':0, \
                        'm_chick_overlap_length':0, 'f_chick_overlap_length':0, \
                        'all3_overlap_length':0, 'mf_bout_overlap_length':0, \
                        'mf_inhale_overlap_length':0, 'mf_coord':0}

length_each = 6/259

for segm in dataset:
    burrow_id = segm[0].split('_')[0]
    label_matrix = segm[-1]
    if burrow_id not in burrow_overlap_dictionary.keys():
        burrow_overlap_dictionary[burrow_id] = overlap_lengths_dict.copy()
    
    chick_inds = np.where(label_matrix[2]==1)[0]
    chick_n = len(chick_inds)
    burrow_overlap_dictionary[burrow_id]['total_chick_length'] +=  chick_n*length_each
    
    male_vec = label_matrix[0] + label_matrix[3] + label_matrix[6]
    male_inds = np.where(male_vec>0)[0]
    male_n = len(male_inds)
    burrow_overlap_dictionary[burrow_id]['total_male_length'] +=  male_n*length_each
    
    female_vec = label_matrix[1] + label_matrix[4] + label_matrix[7]
    female_inds = np.where(female_vec>0)[0]
    female_n = len(female_inds)
    burrow_overlap_dictionary[burrow_id]['total_female_length'] +=  female_n*length_each
    
    m_f_inds = set(male_inds).intersection(set(female_inds))
    m_f_n = len(m_f_inds)
    burrow_overlap_dictionary[burrow_id]['m_f_overlap_length'] +=  m_f_n*length_each

    m_chick_inds = set(male_inds).intersection(set(chick_inds))
    m_chick_n = len(m_chick_inds)
    burrow_overlap_dictionary[burrow_id]['m_chick_overlap_length'] +=  m_chick_n*length_each
    
    f_chick_inds = set(female_inds).intersection(set(chick_inds))
    f_chick_n = len(f_chick_inds)
    burrow_overlap_dictionary[burrow_id]['f_chick_overlap_length'] +=  f_chick_n*length_each
    
    inds_all3 = set(m_chick_inds).intersection(set(female_inds))
    all3_n = len(inds_all3)
    burrow_overlap_dictionary[burrow_id]['all3_overlap_length'] +=  all3_n*length_each

    mf_bouts = label_matrix[3] + label_matrix[4]
    mf_bouts_inds = np.where(mf_bouts==2)[0]
    mf_bout_n = len(mf_bouts_inds)
    burrow_overlap_dictionary[burrow_id]['mf_bout_overlap_length'] +=  mf_bout_n*length_each

    mf_in = label_matrix[0] + label_matrix[1]
    mf_in_inds = np.where(mf_in==2)[0]
    mf_in_n = len(mf_in_inds)
    burrow_overlap_dictionary[burrow_id]['mf_inhale_overlap_length'] +=  mf_in_n*length_each

for burrow in burrow_overlap_dictionary:
    bout_overlap = burrow_overlap_dictionary[burrow]['mf_bout_overlap_length']
    inhale_overlap = burrow_overlap_dictionary[burrow]['mf_inhale_overlap_length']
    if inhale_overlap>0:
        burrow_overlap_dictionary[burrow]['mf_coord'] = inhale_overlap/bout_overlap
    
print(json.dumps(burrow_overlap_dictionary, indent=1)) 
```
